```python
def karşılık_bul(*args, **kwargs):
    for sözcük in args:
        if sözcük in kwargs:
            print("{} = {}".format(sözcük, kwargs[sözcük]))
        else:
            print("{} kelimesi sözlükte yok!".format(sözcük))

sözlük = {"kitap"      : "book",
          "bilgisayar" : "computer",
          "programlama": "programming",
          "beş"        : "five"
         }

karşılık_bul("kitap", "bilgisayar", "programlama", "beş","fonksiyon", **sözlük)
```

# Özyinelemeli (Recursive) Fonksiyonlar

Python fonksiyonları, nasıl başka fonksiyonları çağırabiliyorsa, aynı şekilde, istenirse, kendi kendilerini de çağırabilirler. İşte bu tür fonksiyonlara Python programlama dilinde ‘kendi kendilerini yineleyen’, veya daha teknik bir dille ifade etmek gerekirse ‘özyinelemeli’ (recursive) fonksiyonlar adı verilir.

```python
def fibonacci_of(n):
     if n in {0, 1}:  # Base case
        return n
     return fibonacci_of(n - 1) + fibonacci_of(n - 2)  # Recursive case

print(fibonacci_of(14))
print([fibonacci_of(n) for n in range(20)])
```

```python
def fazalt(s):
    print(s)
    for i in range(1,len(s)):
        print(s[:-i])    # Soldan azalsın [1:]
    
fazalt('Selahattin')
```

```python
def azalt(s):
    if len(s) < 1:
        return s
    else:
        print(s)
        # azalt(s[:-1]) retrn kullanmayabilirsin
        return azalt(s[:-1])
    

print(azalt('Selahattin'))
```

```python
def azalt(s):
    if len(s) < 1:
        return s
    else:
        print('recursive giris :',s)
        azalt(s[1:])
        #raise SystemExit()
        print('recursive giris :',s)

azalt('Selahattin')
```

```python
def ters_çevir(s):
    if len(s) < 1:
        return s
    else:
        print(s[0],end='\t')        
        ters_çevir(s[1:])


ters_çevir('Dursun')
```

```python
def ters_çevir(s):
    if len(s) < 1:
        return s
    else:
        ters_çevir(s[1:])
        print(s[0],end='\t')

ters_çevir('Dursun')
```

```python
def kap_fonk():
    non_local_degisken = 5

    def ic_fonk():
        nonlocal non_local_degisken   # Nerde ileme girecekse orada azılıyor.
        non_local_degisken += 1
        print(non_local_degisken)

    return ic_fonk
```

```python
def islem(sayi, bolen, *eklenenler):
    sonuc = sayi / bolen

    for i in eklenenler:
        sonuc += i

    return sonuc
```

```python
def islem(*eklenenler):
    ekle = 0
    for i in eklenenler:
        ekle += i

    def islem(sayi, bolen):
        return sayi/bolen + ekle

    return islem
```

```python
# Dosya açıp okumayı tekrar olmaması için başka bir fonksiyona ekledik 
def chr_n(karakter_dizisi, karakter):
    sayac = 0
    for i in karakter_dizisi:
        if i == karakter:
            sayac += 1
    return sayac

def file_chr_n(dosya, karakter):
    if type(dosya) == str:
        with open(dosya, "r") as f:
            return chr_n(f.read(), karakter)
    else:
        return chr_n(dosya.read(), karakter)
```

```python
file_chr_n("deco1.txt",'a')
```

```python
# İç içe fonksyion (nested) şeklinde yazdık
def file_chr_n(dosya, karakter):

    def chr_n(karakter_dizisi):
        sayac = 0
        for i in karakter_dizisi:
            if i == karakter:
                sayac += 1
        return sayac

    if type(dosya) == str:
        with open(dosya, "r") as f:
            return chr_n(f.read())
    else:
        return chr_n(dosya.read())
```

```python
file_chr_n("deco1.txt","0")
```

### ‘yield from’ Deyimi

yield from deyimi bir üretecin içinde, başka bir üretecin yield ile döndüreceği değerleri tekrar yield etmek istediğimizde kullanılabilir. 

```python
def gener1():
    yield "gener1 başladı"
    yield "gener1 bitti"

def gener2():
    yield "gener2 başladı"
    yield from gener1()
    yield "gener2 bitti"
```

```python
h= gener2()
for i in range(10):
        print(next(h))
```

```python
for i in gener2():
        print(i)
```

# Modül oluşturma

Pythonda her dosya bir modül olarak adlandırılır. 
Modüller, bazı işlevleri kolaylıkla yerine getirmemizi sağlayan birtakım fonksiyonları ve nitelikleri içinde barındıran araçlardır.  

Python’ın fonksiyon sistemi nasıl bize bir işlevselliği aynı dosya içinde tekrar tekrar kullanma imkanı veriyorsa, modül sistemi de bir fonksiyonu farklı dosyalar ve programlar içinde tekrar tekrar kullanma imkanı verir.

```python
import random
liste = [random.randint(0, 10000) for i in range(10)]
liste
```

Python’da iki farklı fonksiyon türü olduğundan söz etmiştik:

       - Kendi tanımladığımız fonksiyonlar
       - Gömülü (‘built-in’) fonksiyonlar

Aynı şekilde modüller de iki farklı başlık altında incelenebilir:

       - Kendi tanımladığımız modüller
       - Hazır modüller
       
Hazır modüller, Python geliştiricilerinin veya bizim dışımızdaki Python programcılarının yazıp hizmetimize sunduğu modüllerdir. Hazır modüller de kendi içinde ikiye ayrılabilir:

        - Standart Kütüphane Modülleri
        - Üçüncü Şahıs Modülleri
        
Standart Kütüphane Modülleri, doğrudan Python geliştiricileri tarafından yazılıp dile kaynaştırılmış modüllerdir. Bu yönüyle bu modüller gömülü fonksiyonlara çok benzer. Tıpkı gömülü fonksiyonlarda olduğu gibi, Standart Kütüphane Modülleri de her an emrimize amadedir. Biz bunları istediğimiz her an, herhangi bir ek yazılım kurmak zorunda kalmadan, kendi programlarımız içinde kullanabiliriz. Örnek sys, random, math,os

# Decoratörler

```python
def sus_yap(f):    
    def wrapper(*args):
        print("##########################################")
        print("#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#")
        print("#                                        #")
        
        f(*args)              
        
        print("#                                        #")
        print("#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#")
        print("##########################################")
    return wrapper
```

```python
# fonksion dekoratörleri, class dekoratörleri
```

```python
@sus_yap
def den(f):
    print("#{:^40}#".format(f))
```

```python
den("Keramet Subaşı")
```

```python
den("Hasan Gül Sever")
```

```python
def sre_olc(f):
    import time
    def wrapper(*args):
        basl=time.time()
                
        f(*args)
        
        bitis=time.time()
        print("Başlangıç zamanı {}".format(basl))
        print("Bitiş zamanı   {}".format(bitis))
        print(bitis-basl)
    return wrapper
```

```python
@sre_olc
# Faktoriyel hesaplayan fonksiyon Özyinelemeli (Recursive) Fonksiyonlar¶
def faktoriyel(x):
    f=1
    if x==0:
        f=1 
    else:
        for i in range(1,x+1):
            f=f*i 
    return f
```

```python
faktoriyel(227950)
```
